```python
# We need to apply log transformation to reduce skewness

# Exclude 'Total Violent Crimes' and select only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
numeric_cols = numeric_cols.drop("Total Violent Crimes (Cols.3 to 17)", errors='ignore')

# Apply log transformation
df_log = df.copy()
df_log[numeric_cols] = df_log[numeric_cols].apply(lambda x: np.log1p(x))
```

```python
import pandas as pd
import numpy as np

# Check skewness before and after transformation
original_skew = df[numeric_cols].skew()
log_skew = df_log[numeric_cols].skew()

# Create a DataFrame to compare
skewness_comparison = pd.DataFrame({'Original Skewness': original_skew, 'Log Transformed Skewness': log_skew})
print(skewness_comparison)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compare original vs log-transformed data for a highly skewed column
col = "Murder (Sec.302 IPC)"  # Replace with any column of interest

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(df[col], bins=30, kde=True, ax=ax[0])
ax[0].set_title(f"Original Distribution of {col}")

sns.histplot(df_log[col], bins=30, kde=True, ax=ax[1])
ax[1].set_title(f"Log Transformed Distribution of {col}")

plt.show()
```

```python
# As we can see above, curve is more bell-shaped in second graph, meaning less skewed. In above code, we can replace murder with any
# desired column. We can keep transforming the data until the curve is bell shaped.
```

```python
# Remove aggregate rows
df_cleaned = df[~df["State/UT"].isin(["Total State (S)", "Total UT (S)", "Total All India"])]

# Convert all columns from index 3 onwards to numeric (ignoring errors)
df_cleaned.iloc[:, 3:] = df_cleaned.iloc[:, 3:].apply(pd.to_numeric, errors='coerce')

# Sum all crime columns for each state
df_cleaned["Total Crimes"] = df_cleaned.iloc[:, 3:].sum(axis=1)

# Extract relevant data for the pie chart
states = df_cleaned["State/UT"]
crime_counts = df_cleaned["Total Crimes"]

# Create the pie chart
plt.figure(figsize=(17, 17))
plt.pie(crime_counts, labels=states, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})

# Add a title
plt.title("Percentage-wise Distribution of Total Crimes by State")

# Show the plot
plt.show()
```

```python
# Handling outliers using IQR method for each numerical column
for col in df.select_dtypes(include=['number']).columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df = df[(df[col] >= (Q1 - 1.5 * IQR)) & (df[col] <= (Q3 + 1.5 * IQR))]
print("Outliers handled using the IQR method.")
```
